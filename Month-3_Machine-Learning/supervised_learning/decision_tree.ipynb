What is a Decision Tree?

A Decision Tree is a supervised machine learning algorithm used for:

Classification (e.g., spam vs not spam)

Regression (e.g., predicting house prices)

It works by splitting data into branches based on feature conditions until a prediction is made.

Based on Output

| Type                    | Use Case              |
| ----------------------- | --------------------- |
| **Classification Tree** | Output is categorical |
| **Regression Tree**     | Output is continuous  |


How Decision Trees Work (Step-by-Step)

1. Choose the best feature

2. Split data using a condition

3. Repeat recursively

4. Stop when:

    - Max depth reached

    - Pure node

    - Minimum samples reached
